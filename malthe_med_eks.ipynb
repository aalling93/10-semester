# Using functions to plot
Now, we have made a package called malthe with all the functions we made before. This makes it easier when calling and working with.

Initialising by importing your package and others.

```
import malthe_functions as ml
import numpy as np
import matplotlib.pyplot as plt
```

Now, in ml we have all the functions. Lets use them.

```
df = ml.get_df('street_results_99_2.json')
objs,data_obj = ml.get_data(df)
ml.plot_data(data_obj,objs)
#finder distancerne for f√∏rste objekte (dvs. objekt 0). 
dist = ml.get_distances(data_obj[1][1],data_obj[1][2]);
#finder hastigheden..
speed, avg_speed = ml.get_speed(data_obj,1,10)
eta = ml.get_eta(data_obj,1,-5,-8,10)
```

```
place = np.linspace(0, len(speed), num=len(speed))
plt.scatter(place,np.array(speed), s=30, alpha=0.5)
plt.xlabel('')
plt.ylabel('[m/s]')
plt.title('hastighed for id 0. avg speed is:'+str(round(avg_speed,3))+' m/s')
plt.show()


place = np.linspace(0, len(eta), num=len(eta))
plt.scatter(place,np.array(eta), s=30, alpha=0.5)
plt.title('fancy plot of eta')
plt.xlabel('')
plt.ylabel('sekunder')
plt.show()
```
